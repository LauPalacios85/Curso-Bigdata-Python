```python
#Excel
#Indique si hay columnas que borrar (cuales son?)
#Las columnas "Cliente" y "Producto" son redundantes con otras columnas dentro de la base de datos, se verificara si son duplicadas
(dfexcel["Cliente"] == dfexcel["Nombre del cliente"]).all()
```

```python
#Las columnas Cliente y Nombre del cliente son identicas en todas las filas. Se procede a borrar Cliente
dfexcel = dfexcel.drop(columns=["Cliente"])
```

```python
(dfexcel["Producto"] == dfexcel["Nombre del producto"]).all()
```

```python
#Las columnas Producto y Nombre del producto son identicas en todas las filas. Se procede a borra Producto
dfexcel = dfexcel.drop(columns=["Producto"])
```

```python
dfexcel.info()
```

```python
#Realice una copia de su dataframe y en este deje solo los datos que considere necesarios
dfexcel_copia = dfexcel.copy()
```

```python
columnas_necesarias = [
    "Ventas",
    "Año",
    "Mes",
    "Nombre del producto",
    "Nombre del cliente",
    "Empleado",
    "Estado o provincia",
    "País o región",
    "Categoría"
]

dfexcel_copia = dfexcel_copia[columnas_necesarias]
```

```python
dfexcel_copia.info()
```

```python
#Muestre el total de ventas por vendedor
dfexcel.groupby("Empleado")["Ventas"].sum()
```

```python
#Muestre la categoría que más vende y la que menos vende

ventas_categoria = dfexcel.groupby("Categoría")["Ventas"].sum()
ventas_categoria
```

```python
#Categoria que mas vende
ventas_categoria.idxmax()
```

```python
#Categoria que menos vende
ventas_categoria.idxmin()
```
