# CSV Reading


```python
import pandas as pd
df1 = pd.read_csv('felicidad.csv')
print("Primeros 6 registros:")
df1.head(6)
print("Informacion del dataframe:")
df1.info()
print("Tamaño del dataframe:")
df1.shape
print("Numero de filas y columnas:")

num_paises = df1.shape[0]
print(f"Numero de paises: {num_paises}")

# Correcting column names for groupby operations, assuming 'year' and 'Corruption' are the English equivalents.
# Based on the kernel state, 'anio' (year) and 'percepcion_corrupcion' (corruption perception) are the relevant columns.
# If the original intent was 'year' and 'Corruption', those columns might need to be created or mapped from existing ones.
# For now, I will use 'anio' and 'percepcion_corrupcion' from the kernel state.

# Replaced 'year' with 'anio' and 'Corruption' with 'percepcion_corrupcion'
media_corrupcion_anual = df1.groupby("anio")["percepcion_corrupcion"].mean()
print("Media de corrupcion anual:")
print(media_corrupcion_anual)

# Replaced 'year' with 'anio' and 'Corruption' with 'percepcion_corrupcion', and 'Year' with 'anio' for sorting
media_corrupcion_anual = (df1.groupby("anio")["percepcion_corrupcion"].mean().reset_index().sort_values(by="anio"))
print("Media de corrupcion anual:")
print(media_corrupcion_anual)

# Replaced 'Country' with 'pais' and 'Life expectancy' with 'expectativa_vida'
vida_por_pais = df1.groupby("pais")["expectativa_vida"].mean()
print("Expectativa de vida por pais:")
print(vida_por_pais)
```

# Excel Reading

```python
import pandas as pd
df3 = pd.read_excel('base de datos excel tarea sem1.xlsx') # Corrected filename with spaces
df3.head(6)
print("Informacion del dataframe:")
df3.info()
print("Tamaño del dataframe:")
df3.shape
print("Columnas y Filas")


print(df3.columns)
df3_limpio = df3[["Empleado", "Categoría", "Ventas"]].copy() # Corrected 'Vendedor' to 'Empleado'
print(df3_limpio)

# Total de ventas por vendedor
ventas_por_vendedor = (
    df3_limpio.groupby("Empleado")["Ventas"]
    .sum()
    .sort_values(ascending=False)
)

print("Total de ventas por vendedor:")
print(ventas_por_vendedor)

ventas_por_categoria = (
    df3_limpio.groupby("Categoría")["Ventas"]
    .sum()
    .sort_values(ascending=False)
)

print("Ventas por categoría:")
print(ventas_por_categoria)

print("\nCategoría que más vende:")
print(ventas_por_categoria.index[0])

print("\nCategoría que menos vende:")
print(ventas_por_categoria.index[-1])
```
